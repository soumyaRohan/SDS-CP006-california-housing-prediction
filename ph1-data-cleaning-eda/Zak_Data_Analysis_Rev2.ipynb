### Studying Column 'datePostedString' and and trandforming into meaninnful Data

```python
df['postingyear'] = df['datePostedString'].dt.year
df['postingmonth'] = df['datePostedString'].dt.month
df['postingday'] = df['datePostedString'].dt.dayofweek
```

```python
for i in df['postingmonth'].iloc[:,]:
    if i in q1:
        q_list.append('Q1')
    elif i in q2:
        q_list.append('Q2')
    elif i in q3:
        q_list.append('Q3')
    else:
        q_list.append('Q4')
```

### Studying Column 'Levels' and creating more consistent data structure

```python
len(df['levels'].unique())
```

```python
df['levels'].value_counts()
```

```python
one_list =['1','One','One Story','One Story-One','One-Multi/Split','Two Story-One','Three or More Stories-One','Other-One,Multi/Split-One',
           'Other-One']
two_list = ['2','Two','Two Story','Two Story-Two','Two-Multi/Split','Multi/Split-Two','Three or More Stories-Two','Two Story-Three Or More',
           'One-Two','One Story-Two', 'Two-One']
three_list =['3','Three or More Stories','Tri-Level','Three or More Stories-Three Or More','Three Or More-Multi/Split','Three',
             'Tri-Level-Three Or More','Multi/Split-Tri-Level', 'Tri-Level-Two']
threep_list = ['Three Or More','Two-Three Or More','One-Three Or More','Multi/Split-Three Or More','One Story-Three Or More',
               'Three Or More-Split Level','Three Or More-Two','Two-Three Or More-Multi/Split','Three or More Stories-One-Two',
               'One-Two-Three Or More']
four_list = ['4','Four','4+']
five_list =['5','Five or More']
other_list =['Multi/Split','Other','Split Level','Multi/Split-One','One-Two-Multi/Split','Multi-Level']
```

```python
for i, v in df[['levels']].iterrows():
    if v[0] in one_list:
        df['levels'].iloc[i] = '1'
    elif v[0] in two_list:
        df['levels'].iloc[i] = '2'
    elif v[0] in three_list:
        df['levels'].iloc[i] = '3'
    elif v[0] in threep_list:
        df['levels'].iloc[i] = '3+'
    elif v[0] in four_list:
        df['levels'].iloc[i] = '4'
    elif v[0] in five_list:
        df['levels'].iloc[i] = '5'
    elif v[0] in other_list:
        df['levels'].iloc[i] = 'other'
```
