## Dropping Unnecessary Columns, Columns with less unique values and Columns with no significance

```python
df = df.drop(df.columns[0:2], axis=1)
```

### Checking Other Columns, Columns with one unique value to be dropped

```python
df.columns
```

```python
one_col = []
for i in df.columns:
    if len(df[i].unique()) == 1:
        one_col.append(i)
```

```python
for i in one_col:
    print (i,': ',df[i].unique())
```

```python
df = df.drop(one_col, axis=1)
```
